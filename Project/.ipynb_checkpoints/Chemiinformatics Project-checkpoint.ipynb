```python
#!/usr/bin/env python3

# Import relevant libraries to make HTTP requests and parse JSON response
import requests
import json
import pandas as pd

# Set gene_id variable for AR (androgen receptor)
gene_id = "ENSG00000169083"
disease_ids = "MONDO_0004975", ""

# Build query string to get general information about AR and genetic constraint and tractability assessments 
query_string = """query getInfoOnAlzheimers ($efoIds: String!) {
  diseases (efoId: $efoId){
    id
    name
    dbXRefs
    knownDrugs{
      uniqueDrugs
      uniqueTargets
      rows{
        drug{
          id
          name
          tradeNames
          }
          #name
        }
        
      }
      
        }
      }
"""
quer_str = """query target($ensemblId: String!){
    target(ensemblId: $ensemblId){
      id
      approvedSymbol
      biotype
      geneticConstraint {
        constraintType
        exp
        obs
        score
        oe
        oeLower
        oeUpper
      }
      tractability {
        id
        modality
        value
      }
    }
  }
"""

# Set variables object of arguments to be passed to endpoint
variables = {"efoId": disease_id}

# Set base URL of GraphQL API endpoint
base_url = "https://api.platform.opentargets.org/api/v4/graphql"

# Perform POST request and check status code of response
r = requests.post(base_url, json={"query": query_string, "variables": variables})
print(r.status_code)

# Transform API response from JSON into Python dictionary and print in console
api_response = json.loads(r.text)
print(api_response)
new = pd.DataFrame.from_dict(api_response)
print(new)
df = pd.read_json(r.text)
df
new
data = {'drug': {'id': 'CHEMBL807', 'name': 'MEMANTINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL807', 'name': 'MEMANTINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL1699', 'name': 'MEMANTINE HYDROCHLORIDE', 'tradeNames': ['Akatinol', 'Ebixa', 'Marixino', 'Memantine', 'Memantine hydrochloride', 'Namenda', 'Namenda xr', 'Nemdatine', 'Valios']}}, {'drug': {'id': 'CHEMBL3707183', 'name': 'QUINIDINE SULFATE', 'tradeNames': ['Cin-quin', 'Quinicardine', 'Quinidex', 'Quinidine sulfate', 'Quinora']}}, {'drug': {'id': 'CHEMBL807', 'name': 'MEMANTINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL1294', 'name': 'QUINIDINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL1678', 'name': 'DONEPEZIL HYDROCHLORIDE', 'tradeNames': ['Aricept', 'Aricept evess', 'Aricept odt', 'Donepezil hydrochloride']}}, {'drug': {'id': 'CHEMBL502', 'name': 'DONEPEZIL', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL807', 'name': 'MEMANTINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL1699', 'name': 'MEMANTINE HYDROCHLORIDE', 'tradeNames': ['Akatinol', 'Ebixa', 'Marixino', 'Memantine', 'Memantine hydrochloride', 'Namenda', 'Namenda xr', 'Nemdatine', 'Valios']}}, {'drug': {'id': 'CHEMBL1555', 'name': 'GALANTAMINE HYDROBROMIDE', 'tradeNames': ['Galantamine hydrobromide', 'Razadyne', 'Razadyne er']}}, {'drug': {'id': 'CHEMBL636', 'name': 'RIVASTIGMINE', 'tradeNames': ['Alzest', 'Eluden', 'Erastig', 'Exelon', 'Kerstipon', 'Nimvastid', 'Prometax', 'Rivastigmine', 'Somniton', 'Voleze']}}, {'drug': {'id': 'CHEMBL723', 'name': 'CARVEDILOL', 'tradeNames': ['Carvedilol', 'Coreg', 'Coreg CR', 'Eucardic 12.5', 'Eucardic 25', 'Eucardic 3.125', 'Eucardic 6.25']}}, {'drug': {'id': 'CHEMBL502', 'name': 'DONEPEZIL', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL807', 'name': 'MEMANTINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL1699', 'name': 'MEMANTINE HYDROCHLORIDE', 'tradeNames': ['Akatinol', 'Ebixa', 'Marixino', 'Memantine', 'Memantine hydrochloride', 'Namenda', 'Namenda xr', 'Nemdatine', 'Valios']}}, {'drug': {'id': 'CHEMBL3707183', 'name': 'QUINIDINE SULFATE', 'tradeNames': ['Cin-quin', 'Quinicardine', 'Quinidex', 'Quinidine sulfate', 'Quinora']}}, {'drug': {'id': 'CHEMBL636', 'name': 'RIVASTIGMINE', 'tradeNames': ['Alzest', 'Eluden', 'Erastig', 'Exelon', 'Kerstipon', 'Nimvastid', 'Prometax', 'Rivastigmine', 'Somniton', 'Voleze']}}, {'drug': {'id': 'CHEMBL112', 'name': 'ACETAMINOPHEN', 'tradeNames': ['Abdine', 'Acephen', 'Acetaminophen', 'Alvedon', 'Angiers', 'Calpol', 'Calpol six plus', 'Calpol six plus fastmelts', 'Child lemsip', 'Cupanol', 'Dafalgan', 'Datril', 'Disprol', 'Disprol infant', 'Disprol jnr', 'Disprol paed', 'Fennings', 'Galpamol', 'Hedex', 'Infadrops', "Infants' feverall", 'Injectapap', 'Junior parapaed', 'Mandanol', 'Medinol', 'Medinol for children', 'Medinol over 6', 'Medinol under 6', 'Medised plain', 'Miradol', 'Neopap', 'Ofirmev', 'Paldesic', 'Panadol', 'Panadol actifast', 'Panadol actifast solb', 'Panadol advan', 'Panadol jnr', 'Panadol oa', 'Panaleve', 'Panaleve 6 plus', 'Paracetamol', 'Paramin', 'Parapaed jnr', 'Parapaed six plus', 'Paravict', 'Perfalgan', 'Phenaphen', 'Placidex', 'Rimadol', 'Salzone', 'Tixymol', 'Tramil 500', 'Tylenol']}}, {'drug': {'id': 'CHEMBL807', 'name': 'MEMANTINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL3', 'name': 'NICOTINE', 'tradeNames': ['Habitrol', 'Nicabate', 'Nicoderm', 'Nicoderm cq', 'Nicopass', 'Nicopatch', 'Nicorette invisi', 'Nicotine', 'Nicotinell classic', 'Nicotinell support', 'Nicotinell tts 10', 'Nicotinell tts 20', 'Nicotinell tts 30', 'Nicotrol', 'Nicotrol Inhaler', 'Nicotrol NS', 'Niquitin', 'Niquitin clr', 'Niquitin minis cherry', 'Niquitin minis mint', 'Niquitin minis orange', 'Niquitin mint', 'Niquitin pre-quit', 'Niquitin strips mint', 'Prostep', 'Stoppers', 'Stubit']}}, {'drug': {'id': 'CHEMBL807', 'name': 'MEMANTINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL1699', 'name': 'MEMANTINE HYDROCHLORIDE', 'tradeNames': ['Akatinol', 'Ebixa', 'Marixino', 'Memantine', 'Memantine hydrochloride', 'Namenda', 'Namenda xr', 'Nemdatine', 'Valios']}}, {'drug': {'id': 'CHEMBL1294', 'name': 'QUINIDINE', 'tradeNames': []}}, {'drug': {'id': 'CHEMBL1699', 'name': 'MEMANTINE HYDROCHLORIDE', 'tradeNames': ['Akatinol', 'Ebixa', 'Marixino', 'Memantine', 'Memantine hydrochloride', 'Namenda', 'Namenda xr', 'Nemdatine', 'Valios']}}
df = pd.DataFrame.from_dict(data)
df
#pd.read
```

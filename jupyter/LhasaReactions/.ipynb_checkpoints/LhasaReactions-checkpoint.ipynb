# Resurrecting the Lhasa Reaction Knowledge Base - the SAVI project.

Lhasa was originally a retrosynthesis software. We have re-implemented a language intepreter for CHMTRN/PATRAN, the knowledge base specification languages, in a clean-room effort. 

We can use both original Lhasa knowledge base entries, and encode new reaction knowledge developed after the 80ies.

The software is operated in a scripting environment and is designed to be adapted to different tasks. SAVI uses the engine for forward reaction prediction, generating a humungous database of reaction products which are synthesizable in a single step from commercially available starting materials.

From a Lhasa transform, a couple of basic operations are derived:
1. Starting material scans
2. Forward reactionsd
3. Retrosynthetic scoring
4. Report generation

We will demonstrate these steps, on a very simple reaction, and highlight why these are much more powerful than the run-of-the mill transform pattern application in standard library enumerators, especially with respect to the reliability of the proposed reactions.


It does not get much more simple:

## Williamson Ether Synthesis

from the original Lhasa transform set, but slightly recoded and updated by P. Judson.

It does not get much simpler. https://en.wikipedia.org/wiki/Williamson_ether_synthesis.

Do you think this can be conclusively solved with a trivial SMIRKS pattern?

```python
Prop.Setparam('E_SVG_IMAGE','frame',False)
from IPython.display import HTML,SVG,Image
print(cactvs['data_directory'])
```

```python

smirks=Reaction('[C;z1:1][O:2][#1].[Cl,Br,I][C;z1:3]>>[C:1][O:2][C:3]','smirks')
reagents=Ens('CCO.ClC')
products=reagents.transform(smirks,direction='forward',reactionmode='singlestep')
print(products)
display(SVG(products[0].E_SVG_IMAGE))
```

Essentially, that is what all these naive transform-based enumerators do. How does this trivial case look like with the Lhasa engine?

We being by loading a byte-compiled version of the Williamson Ether synthesis rule, and query a few attributes.

```python
l=Lhasa.Read(6039)
print(l.name)
print(l.version)
print(l.author)
print(l.sourcefiles)
print(l.typicalyield)
print(l.reputation)
print(l.generalrefs)
```

React a starting material ensemble in forward direction. We get a duplicate-filtered list of products which can be formed according to the simple reaction pattern in the transform code. In this case, there is only a single product.

```python
reagents=Ens('CCO.ClC')
products = l.react(reagents)
print(products)
display(SVG(products[0].E_SVG_IMAGE))
```

This looks very much the same as the SMIRKS approach so far. Where is the intelligence?

The intelligent part is in the retrosynthetic scoring of the forward products.

```python
score=l.score(products[0])
print(score)
```

Do you have to believe that?

No, there is an explanation.

```python
print(l.scoredreactions)
r1=l.scoredreactions[0]
r2=l.scoredreactions[1]
r3=l.scoredreactions[3]
print(r1)
```

There a multiple retrosynthetic candidate retro-reactions, each with its own score. The overall score is just the best score of any of those. So what are these candidate reactions?

```python
display(Image(r1.X_GIF))
```

```python
display(Image(r2.X_GIF))
```

```python
display(Image(r3.X_GIF))
```

Hmm. Which one of the scored alternatives corresponds to the exact reaction we used in the forward forward reaction?

(*) The algorithm shown only works for standard cases. Like this one. Some transforms get more fancy, like attaching superatoms, and then we need to use more advanced substructure matching to determine the mnatch between retro reaction and forward reaction.

```python
print(reagents.E_STEREO_HASHY)
for x in l.scoredreactions:
    if (x.reagents.E_STEREO_HASHY == reagents.E_STEREO_HASHY):
        print(x)
```

By the way, what happens to the scoring for symmetrical compounds?

```python
l.score('CCOCC')
print(l.scoredreactions)
```

Results are automatically deduplicated (plus some other filter magic) - just three reactions (Cl,Br,I) here.

Let's observe the effects of a few non-participating functional groups on different retrosynthetic scores.

Amines are not a death sentence, but lower the score considerably (66)

```python
print(l.score('NCCCOC'))
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

What happens if you work with a phenol instead of a simple alkyl alcohol?

```python
print(l.score(Ens('c1ccccc1OC')))
print(l.scoredreactions[0].X_LHASA_SCORE)
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

In that case, you can now use alternative, more gentle reaction conditions because phenols are more acidicm, and that is annotated.

What about other alcohols in the structure?

```python
l.score('OC(C)CCCOCC')
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

The reaction gets seriously downgraded because this is not a symmetric case.

What about stereochemistry?

The transform is stereospecific, and if there is stereochemistry at the reaction center, it gets inverted (92)

```python
print(l.ruleflags)
l.score('C[C@H](CC)OC')
display(HTML(filename=l.scoredreactions[3].X_LHASA_REPORT))
```

So far for a few highlights in the top-level transform code. But what happens in the functions implemented in the various auxiliary libray files?

```python
print(l.sourcefiles)
```

The library part is a generic inspection engine for SN2 substitutions which can detect what can go awry with these. It is reusable for all reactions of that class. Some parts of the library code are never executed in this transform because here we have always an implicit hard/basic reaction condition setting. For the generic SN2 checks, there are about 280 LOC total - larger than the actual top-level Williams transform.

Let's have a look at some of the evaluations performed in these code sections, and understand some emergency kills issued therein.

```python
# No SN2 on vinyl Cl (or any other ligand) (20)
print(l.score('COC=C'))
print(l.scoredreactions)
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

```python
# No SN2 on tertiary C if not intramolecular (26)
# Having the O on the tert center is OK, though
print(l.score('COC(C)(C)C'))
print(l.scoredreactions)
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

```python
# But in the intramolecular case, this actually works
print(l.score('C1OC(C)(C)CCCC1'))
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

```python
# C*3 in ring of size 6, but C*2 not in the same ring - this eliminates if there is
# hydrogen on the alpha atom (53)
# You can not have the Cl on the ring. The other way around works.
print(l.score('COC1CCCCC1'))
print(l.scoredreactions)
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

```python
# But this works again because there is no chance for in-ring elimination
print(l.score('COC1C(C)(C)CCC=C1'))
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

```python
# Steric hindrance in various variants
print(l.score('COC(C(C)(C)C)C(C)=C'))
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

```python
# Ring sizes
print(l.score('C1OCC1'))
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

OK, so far this was all straightforward chemistry. 

But what happens if a simple SN2 substitution begins to behave erratically and mutates into SN2'?

Remember: SN2' is attack of the nucleophile on the remote end of a double bond, with the leaving bond in allylic position, and the double bond shifting one position in the product.

```python
# OK, this is probably still safe, no hydrogen on the remote side, direct attack favored (50)
print(l.score('COC(C)C=C(C)C'))
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

```python
# As is this one - SN' attack happens, but the results are equivalent (65)
print(l.score('COC(C)C=CC'))
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

And what about rings?

```python
# Postmechanism check bridgeheads (12)
print(l.score('C12CCC(CC1)(C2)OC'))
print(l.scoredreactions)
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

```python
# Postmechanism check 3-ring - cannot perform the substitution on the 3-ring carbon (13)
print(l.score('COC1CC1'))
print(l.scoredreactions)
display(HTML(filename=l.scoredreactions[0].X_LHASA_REPORT))
```

These were the basics of how to run forward reactions, how to score these, and how to check that
the scored set both actually contains the forward reaction, and has an acceptable score.

But how do we find suitable starting materials for these transforms if we want to batch-generate libraries of synthesizable molecules?

The Lhasa object supports the generation of queries for reagents from the patterns - for inter- or intramolecular reactions, and with constraints for the multiplicity of potentially reacting groups.

We set up an in-memory sample dataset for demonstration - in the standard case, these
queries would be run on an SD file, or in our database cartridge.

```python
d=Dataset('CCCl','CCBr','ClCCCl','CCO','CCCO','OCCCO','ClCCCCO','C1COCCC1CO','CCC(=O)O','CC(Cl)Cl')
display(SVG(d.get('D_SVG_IMAGE',parameters={'ncols':4,'nrows':'3'})))
```

```python
q1=l.intermol_reagentquery1
print(q1)
print(d.scan(q1,mode='recordlist'))
```

```python
print(d.scan(l.intermol_reagentquery2,mode='recordlist'))
```

```python
print(d.scan(l.intramol_reagentquery,mode='recordlist'))
```

```python
print(d.scan(l.reagentquery1,mode='recordlist'))
```
